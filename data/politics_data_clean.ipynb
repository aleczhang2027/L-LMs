```python

"""
politics_data_clean.ipynb

Whole twitter dataset from: https://www.kaggle.com/datasets/prkhrawsthi/twitter-sentiment-dataset-3-million-labelled-rows

Download as twitter_dataet.csv and run this script to get the clean politics dataset.
Add or delete keywords as needed to better match the political topics, entities, or terms you want to capture.
"""

import re
import pandas as pd

# ---- files ----
INPUT_FILE = "twitter_dataset.csv"
OUTPUT_FILE = "politics_sentiment.csv"

# ---- political keywords: single words only ----
keywords = {
    "election",
    "democrat", "democrats",
    "republican", "republicans",
    "biden", "trump", "obama",
    "senate", "congress",
    "president",
    "government",
    "politics", "political",
    "bureaucracy",
    "minister",
    "governor",
    "senator",

    # ideology words
    "capitalism",
    "socialism",
    "communism",
    "fascism",
    "libertarian",
    "neoliberal",
    "nationalism",
    "populism",
    "authoritarian",

    # policy topic words
    "immigration",
    "taxation",
    "tariffs",
    "subsidy", "subsidies",
    "inflation",
    "budget",
    "deficit",
    "welfare",
    "healthcare",
    "medicare",
    "medicaid",
    "unemployment",
    "regulation",
    "deregulation"
}

# ---- build one regex that matches whole words only ----
pattern = re.compile(r"\b(?:{})\b".format("|".join(re.escape(word) for word in keywords)), re.IGNORECASE)

# ---- read csv ----
df = pd.read_csv(INPUT_FILE)

# ---- keep only rows where tweet contains at least one political keyword ----
political_df = df[df["tweet"].astype(str).str.contains(pattern, na=False)].copy()

# ---- save result ----
political_df.to_csv(OUTPUT_FILE, index=False)

# ---- print summary ----
print(f"Original rows: {len(df):,}")
print(f"Political rows kept: {len(political_df):,}")
print(f"Saved to: {OUTPUT_FILE}")
```
